```python
import re
import warnings
warnings.filterwarnings("ignore")
# find element ids inside column Name from dataframe df
# eg: Seating-LAMMHULTS-PENNE-Chair [289790]  => 289790
pattern = r'\[(.*?)\]'
element_ids = df["Name"].str.findall(pattern)
element_ids = element_ids.str[0].str.replace(r'\[|\]', '')
if "ElementID" in df.columns:
    df["ElementID"] = element_ids
else:
    df.insert(2, "ElementID", element_ids)
df
```

```python
# now we know how many element ids in the dataframe
elements = df["ElementID"].unique()
# remove all nan
elements = elements[~pd.isnull(elements)]
elements = elements.astype(int)
print("Total Element IDs: ", len(elements))
```
